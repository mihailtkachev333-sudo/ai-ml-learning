```python
def age_group(age):
    if pd.isna(age):
        return 'Unknown'
    elif age < 18:
        return 'Child'
    elif age <= 60:
        return 'Adult'
    else:
        return 'Senior'

df['AgeGroup'] = df['Age'].apply(age_group)
print(df['AgeGroup'].value_counts())
```

```python
print(df.groupby('AgeGroup')["Survived"].mean())
```

```python
print(df.groupby("Embarked")["Survived"].mean().max())
```
